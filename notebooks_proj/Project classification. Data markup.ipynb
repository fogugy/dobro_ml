```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_src = pd.read_csv('./data/projects_done.csv', sep=';')
```

```python
try:
    df = pd.read_csv('./data/data_done.csv', sep=';')
except:
    df = data_src.copy()
    df['common'] = [-1]*df.shape[0]
```

```python
df[df['common'] == -1].head()
```

```python
for index, row in df.iterrows():
    if df.at[index, 'common'] != -1:
        continue
    
    print("-"*10)
    print(row['title'])
    print(row['short'])
    
    if index % 10 == 0:
        print(f'{index}/{df.shape[0]}')
        df.to_csv('./data/data_done.csv', index=False, sep=';')
    
    a = input()
    df.at[index,'common'] = a or '0'
```

```python
df.to_csv('./data/data_done.csv', index=False, sep=';')
```
